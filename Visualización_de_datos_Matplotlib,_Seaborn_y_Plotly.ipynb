#Visualización de datos: creando gráficos con bibliotecas de Python: Matplotlib, Seaborn y Plotly

##1. Colombia

```python
import pandas as pd
df = pd.read_csv('inmigrantes_canada.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.set_index('Pais', inplace=True)
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
colombia = df.loc['Colombia', anos]
colombia
```

```python
col_dict = {'Año': colombia.index.tolist(),
            'Inmigrantes':colombia.values.tolist()}
col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.head()
```

##Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
#Manual
plt.figure(figsize=(8, 4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.title('Inmigración de colombianos a Canadá')
plt.xlabel('Año')
plt.ylabel('Inmigrantes')
plt.show()
```

```python
#Con variables
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigración de colombianos a Canadá\n1980 - 2013')
ax.set_xlabel('Año')
ax.set_ylabel('Número de Inmigrantes')
plt.show()
```

```python
fis, axs = plt.subplots(1, 2, figsize=(10, 3))

axs[0].plot(datos_col['Año'], datos_col['Inmigrantes'])
axs[0].set_title('Inmigración de colombianos a Canadá\n1980 - 2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Número de Inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('Boxplot de la inmigración de colombianos a Canadá')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('Número de Inmigrantes')
axs[1].grid()

plt.style.use('fivethirtyeight')
plt.show()
```

```python
datos_col.describe()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
#Evita superposición de gráficos
fig.subplots_adjust(hspace=0.5, wspace=0.3)

#Recuerda que el valor pasado a estos parámetros es un número decimal
#que representa la fracción del tamaño de la figura, por ejemplo, 0.5
#significa que el espaciado será del 50% de la altura/anchura de la figura.

fig.suptitle('Inmigración de los 4 mayores países sudamericanos\nhacia Canadá 1980 - 2013')

axs[0,0].plot(df.loc['Colombia', anos])
axs[0,0].set_title('Colombia')

axs[0,1].plot(df.loc['Brasil', anos])
axs[0,1].set_title('Brasil')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Perú', anos])
axs[1,1].set_title('Perú')

ymin = 0
ymax = 7000

#Buena práctica: mantener escala de los ejes.
for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

for ax in axs.flat:
  ax.set(xlabel='Año', ylabel='Número de Inmigrantes')
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.grid()

plt.show()
```

```python
#Con variables
fig, ax = plt.subplots(figsize=(8, 3))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'], lw=2.5, color='slateblue') #, marker='o', markersize=2.5)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigración de colombianos a Canadá\n1980 - 2013', fontsize=18, loc='left')
ax.set_xlabel('Año', fontsize=14)
ax.set_ylabel('Número de Inmigrantes', fontsize=14)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)
plt.grid(linestyle='--')
plt.show()
```

```python
colores = []
for pais in sudamerica_sorted.index:
  if pais == 'Colombia':
    colores.append('forestgreen')
  else:
    colores.append('silver')

fig, ax = plt.subplots(figsize=(12, 5))

ax.set_frame_on(False)
ax.xaxis.set_visible(False)
ax.tick_params(axis='y', length=0)

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Sudamérica: Colombia es el segundo país con más migrantes\nhacia Canadá durante el periodo 1980 - 2013',
             fontsize=16, loc='left')
ax.set_xlabel('Número de Inmigrantes')
ax.xaxis.set_tick_params(labelsize=12)
plt.grid(linestyle='--')

for i,j in enumerate(sudamerica_sorted['Total']):
  ax.text(j+20, i, str(j), color='black', fontsize=8.5, ha='left', va='center')

fig.savefig('/content/inmigrantes_sudamericanos_canada.png',
            transparent=False, dpi=300, bbox_inches='tight')
plt.show()
```

##Seaborn

```python
import seaborn as sns
```

```python
sns.set_theme()
```

```python
top_10 = df.sort_values('Total', ascending=False).head(10)
top_10
```

```python
ax = sns.barplot(data=top_10, x='Total', y=top_10.index, orient='h')
ax.set(title='Top 10 países con mayor migración hacia Canadá\n1980 - 2013',
       xlabel='Número de migrantes',
       ylabel='Año')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))

ax = sns.barplot(data=top_10, x='Total', y=top_10.index, orient='h')
ax.set_title('Top 10 países con mayor migración hacia Canadá\n1980 - 2013',
       loc='left', fontsize=16)

ax.set_xlabel('Número de Inmigrantes', fontsize=14)
ax.set_ylabel('')

plt.show()
```

```python
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8, 4))
  ax.set_frame_on(False)
  #ax.xaxis.set_visible(False)
  #ax.tick_params(axis='y', which='both', size=0)
  ax = sns.barplot(data=top_10, x='Total', y=top_10.index, hue=top_10.index, orient='h',
                   palette=palette, legend=False)
  ax.set_title('Top 10 países con mayor migración hacia Canadá\n1980 - 2013',
        loc='left', fontsize=16)
  ax.set_xlabel('Número de Inmigrantes', fontsize=14)
  ax.set_ylabel('')
  for i,j in enumerate(top_10['Total']):
    ax.text(j+20, i, str(j), color='black', fontsize=8.5, ha='left', va='center')
  plt.show()
```

```python
generar_grafico('rocket')
```

```python
generar_grafico('cubehelix')
```

```python
def generar_grafico2(palette):
  fig, ax = plt.subplots(figsize=(12, 6))
  ax.set(xticklabels=[])
  ax = sns.barplot(data=top_10, x='Total', y=top_10.index, hue=top_10.index, orient='h',
                   palette=palette, legend=False)
  ax.set_title('Top 10 países con mayor migración hacia Canadá\n1980 - 2013',
        loc='left', fontsize=16)
  ax.set_xlabel('Número de Inmigrantes', fontsize=14)
  ax.set_ylabel('')
  sns.despine()
  for i,j in enumerate(top_10['Total']):
    ax.text(j+20, i, str(j), color='black', fontsize=8.5, ha='left', va='center')
  #This line saves the file
  plt.savefig(f'grafico_{palette}.png', dpi=300, bbox_inches='tight')
  plt.show()
```

```python
sns.set_theme(style='darkgrid')
generar_grafico2('rocket')
```

```python
sns.set_theme(style='whitegrid')
generar_grafico2('rocket')
```

```python
sns.set_theme(style='white')
generar_grafico2('rocket')
```

##2. Comparación Brasil vs Argentina

```python
df_comparacion = df.loc[['Brasil', 'Argentina'], anos]
df_comparacion.head()
```

```python
df_comparacion = df_comparacion.T
df_comparacion.head()
```

```python
#Para hacer comparaciones puedo colocar dos líneas con plt.plot()
plt.plot(df_comparacion['Brasil'], label='Brasil')
plt.plot(df_comparacion['Argentina'], label='Argentina')
plt.title('Inmigración de Brasil y Argentina a Canadá')
plt.xlabel('Año')
plt.ylabel('Número de Inmigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend()
plt.grid()
plt.show()
```
